 Data Cleaning

```python
#NULL values 
df.isnull().sum()
```

```python
#removing NULL values
new_data= df.dropna()
```

```python
new_data.isnull().sum()
```

```python
new_data.head()
```

```python
#converting Name and Surname columns into str
new_data['Name'] = new_data['Name'].astype(str)
new_data['Surname'] = new_data['Surname'].astype(str)
```

```python
#concatenation of two columns
new_data['Full Name']= new_data['Name']+ '' + new_data['Surname']

def Names(row):
     return row['Name']+ '' + row['Surname']

new_data['Full Name']= new_data.apply(Names, axis=1)
new_data.head()
```

```python
new_data.head()
```

```python
#Extract the month from the date
new_data['Date'] = pd.to_datetime(new_data['Date'])

new_data['Month'] = new_data['Date'].dt.month
new_data.head()
```

```python
#Total Monthly Transactions
results= new_data.groupby('Month').sum()
results
```

```python
#bar plot on months vs Transactions 
months =range(1,13)

plt.plot(months, results['Transaction Amount'])
plt.xticks(months)
plt.xlabel('month numer')
plt.ylabel('Transactions')
plt.show()
```

```python
#Gender Transactions
results= new_data.groupby('Gender').sum()
results
```

Exploratory Data Analysis-(EDA)

```python
# plotting a bar chart for Gender vs Category Count
ax = sns.countplot(x = 'Gender',data = new_data, hue='Category')
sns.set(rc={'figure.figsize':(30,15)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#category wise count 
ax = sns.countplot(x = 'Category',data = new_data)
sns.set(rc={'figure.figsize':(15,10)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# plotting a bar chart for Date vs Top 10 Transaction Amounts      
sales_data = new_data.groupby(['Date'], as_index=False)['Transaction Amount'].sum().sort_values(by='Transaction Amount', ascending=False).head(10)
sns.set(rc={'figure.figsize':(40,20)})
sns.barplot(data = sales_data, x = 'Date',y= 'Transaction Amount')
```

```python
# plotting a bar chart for Date vs Buttom 10 Transaction Amounts      
sales_data = new_data.groupby(['Date'], as_index=False)['Transaction Amount'].sum().sort_values(by='Transaction Amount', ascending=False).tail(10)
sns.set(rc={'figure.figsize':(40,20)})
sns.barplot(data = sales_data, x = 'Date',y= 'Transaction Amount')
```

By this customer Transaction Analysis, Females made the highest number of transactions.The Travel category records the most transactions, May month having the highest transaction volume throughout the year and the lowest number of transactions occurred in November month.
